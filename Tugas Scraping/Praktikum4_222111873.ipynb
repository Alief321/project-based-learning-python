## Web Scraping Yahoo!Finance dengan Request + BeautifulSoap

### Scraping data

```python
import requests
from bs4 import BeautifulSoup

# Mendapatkan page HTML melalui request
i = 0
hasil_data = []

for i in range(0, 75, 25):
    halaman = requests.get('https://finance.yahoo.com/crypto/?count=25&offset='+ str(i),
                            headers={'User-Agent': 'Mozilla/5.0'})

    # Parsing konten halaman menggunakan beautifulsoup
    halaman_parsed = BeautifulSoup(halaman.content, 'html.parser')
    finances = halaman_parsed.findAll('tr', {'class': 'simpTblRow'})

    for finance in finances:
        symbol = finance.select_one('td a').get_text()
        img_symbol = finance.select_one('td img')['src']
        nama = finance.select('td')[1].get_text()
        harga = finance.select_one('td:nth-child(3)').get_text()
        perubahan = finance.select_one('td:nth-child(4)').get_text()
        persen_perubahan = finance.select_one('td:nth-child(5)').get_text()
        nilai_pasar = finance.select_one('td:nth-child(6)').get_text()
        volume_in_currency_kumulatif = finance.select_one('td:nth-child(7)').get_text()
        volume_in_currency_24hr = finance.select_one('td:nth-child(8)').get_text()
        Total_volume_all_currencies = finance.select_one('td:nth-child(9)').get_text()
        circulating_supply = finance.select_one('td:nth-child(10)').get_text()

        hasil_data.append({"symbol": symbol, "img_symbol" : img_symbol, "nama": nama, "harga": harga, "perubahan": perubahan, "persen_perubahan": persen_perubahan, "nilai_pasar": nilai_pasar, "volume_in_currency_kumulatif": volume_in_currency_kumulatif, "volume_in_currency_24hr": volume_in_currency_24hr, "Total_volume_all_currencies": Total_volume_all_currencies, "circulating_supply": circulating_supply})

print(hasil_data)
```

### Mengimport ke csv

```python
import pandas as pd

hasil_data_df = pd.DataFrame(hasil_data)
hasil_data_df.to_csv('./content/yahoo-finance.csv', index=False)
```
